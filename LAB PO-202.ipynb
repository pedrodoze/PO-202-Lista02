```python
import pulp
import math

prob1 = pulp.LpProblem('Problema 1', pulp.LpMinimize)

x1 = pulp.LpVariable('Milho', lowBound=0, cat='Continuous')
x2 = pulp.LpVariable('Soja', lowBound=0, cat='Continuous')

custo = 0.26*x1 + 0.32*x2

prob1 += custo

proteina = 0.07*x1 + 0.21*x2

prob1 += (proteina>=0.34)

carbo = 0.82*x1 + 0.79*x2

prob1 += (carbo>=2.64)

print (prob1)

optimization_result = prob1.solve()

assert optimization_result == pulp.LpStatusOptimal

for var in (x1,x2):
    print(' A quantidade ótima de {} é: {:1.3f}'.format(var.name, var.value()))
```

```python
prob = pulp.LpProblem('Problema 2', pulp.LpMaximize)

x1 = pulp.LpVariable('x2', lowBound=0, cat='Integer')
x2 = pulp.LpVariable('x3', lowBound=0, cat='Integer')

max = (math.log(1.17)-2*math.log(1.08))*x1 + (math.log(1.27)-3*math.log(1.08))*x2

prob += max

rest1 = 5 - 2*x1 - 3*x2

prob += (rest1>=0)

print (prob)

optimization_result = prob.solve()

assert optimization_result == pulp.LpStatusOptimal

for var in (x1,x2):
    print(' A quantidade ótima de {} é: {:1.3f}'.format(var.name, var.value()))
```
